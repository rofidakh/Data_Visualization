```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
allfood_df = pd.read_csv(r"C:\Users\Dell_i7\Desktop\visual_project\food\NutritionalFacts_Fruit_Vegetables_Seafood.csv", encoding='latin1')
```

```python
# Filter vegetables and fruits
vegetables_and_fruits_df = allfood_df[(allfood_df["Food Type"] == "Vegetables") | (allfood_df["Food Type"] == "Fruits")]
```

```python
# Filter seafood
seafood_df = allfood_df[allfood_df["Food Type"] == "Seafood"]
```

```python
# List of key nutrients
key_nutrients = ['Calories', 'Total Fat', 'Sodium', 'Potassium', 'Total Carbo-hydrate',
                 'Dietary Fiber', 'Protein', 'Vitamin A', 'Vitamin C', 'Calcium',
                 'Iron', 'Saturated Fat', 'Chole-sterol_1']
```

```python
# Convert nutrient columns to numeric
for nutrient in key_nutrients:
    vegetables_and_fruits_df[nutrient] = pd.to_numeric(vegetables_and_fruits_df[nutrient], errors='coerce')
    seafood_df[nutrient] = pd.to_numeric(seafood_df[nutrient], errors='coerce')
```

```python
# Create a figure
plt.figure(figsize=(30, 35))

# Density distribution plots for each nutrient
for i, nutrient in enumerate(key_nutrients):
    plt.subplot(6, 3, i + 1)
    sns.kdeplot(vegetables_and_fruits_df[nutrient], color='green', label='Vegetables')
    sns.kdeplot(seafood_df[nutrient], color='blue', label='Seafood')
    plt.title(nutrient)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.xlim(left=0)  # Set lower limit of x-axis to zero
```

```python
# Scatter plot for relationship between calories and protein
# Create a figure
plt.figure(figsize=(30, 35))
plt.subplot(6, 3, len(key_nutrients) + 1)
plt.scatter(vegetables_and_fruits_df['Calories'], vegetables_and_fruits_df['Protein'], color='green', label='Vegetables')
plt.scatter(seafood_df['Calories'], seafood_df['Protein'], color='blue', label='Seafood')
plt.title('Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.legend()
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for relationship between calories and protein
# Create a figure
plt.figure(figsize=(30, 35))
plt.subplot(6, 3, len(key_nutrients) + 1)
plt.scatter(vegetables_and_fruits_df['Calories'], vegetables_and_fruits_df['Total Fat'], color='green', label='Vegetables')
plt.scatter(seafood_df['Calories'], seafood_df['Total Fat'], color='blue', label='Seafood')
plt.title('Calories vs Total Fat')
plt.xlabel('Calories')
plt.ylabel('Total Fat')
plt.legend()
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for relationship between calories and protein
# Create a figure
plt.figure(figsize=(30, 35))
plt.subplot(6, 3, len(key_nutrients) + 1)
plt.scatter(vegetables_and_fruits_df['Calories'], vegetables_and_fruits_df['Sodium'], color='green', label='Vegetables')
plt.scatter(seafood_df['Calories'], seafood_df['Sodium'], color='blue', label='Seafood')
plt.title('Calories vs Sodium')
plt.xlabel('Calories')
plt.ylabel('Sodium')
plt.legend()
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for relationship between calories and protein
# Create a figure
plt.figure(figsize=(30, 35))
plt.subplot(6, 3, len(key_nutrients) + 1)
plt.scatter(vegetables_and_fruits_df['Calories'], vegetables_and_fruits_df['Potassium'], color='green', label='Vegetables')
plt.scatter(seafood_df['Calories'], seafood_df['Potassium'], color='blue', label='Seafood')
plt.title('Calories vs Potassium')
plt.xlabel('Calories')
plt.ylabel('Potassium')
plt.legend()
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for relationship between calories and protein
# Create a figure
plt.figure(figsize=(30, 35))
plt.subplot(6, 3, len(key_nutrients) + 1)
plt.scatter(vegetables_and_fruits_df['Calories'], vegetables_and_fruits_df['Total Carbo-hydrate'], color='green', label='Vegetables')
plt.scatter(seafood_df['Calories'], seafood_df['Total Carbo-hydrate'], color='blue', label='Seafood')
plt.title('Calories vs Total Carbo-hydrate')
plt.xlabel('Calories')
plt.ylabel('Total Carbo-hydrate')
plt.legend()
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# Calculate correlations for vegetables and fruits
correlations_vegetables_fruits = {}
for nutrient in key_nutrients:
    if nutrient == "Calories" :
        continue
    correlation = vegetables_and_fruits_df[nutrient].corr(vegetables_and_fruits_df['Calories'])
    correlations_vegetables_fruits[nutrient] = correlation

# Calculate correlations for seafood
correlations_seafood = {}
for nutrient in key_nutrients:
    if nutrient == "Calories" :
        continue
    correlation = seafood_df[nutrient].corr(seafood_df['Calories'])
    correlations_seafood[nutrient] = correlation

# Output correlations
print("Correlations for Vegetables and Fruits:")
for nutrient, correlation in correlations_vegetables_fruits.items():
    print(f"{nutrient}: {correlation}")

print("\nCorrelations for Seafood:")
for nutrient, correlation in correlations_seafood.items():
    print(f"{nutrient}: {correlation}")

# Scatter plot for correlations with calories for vegetables and fruits
plt.figure(figsize=(10, 5))
plt.scatter(correlations_vegetables_fruits.keys(), correlations_vegetables_fruits.values(), color='green', label='Vegetables and Fruits')
plt.title('Correlation with Calories for Vegetables and Fruits')
plt.xlabel('Nutrient')
plt.ylabel('Correlation with Calories')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot for correlations with calories for seafood
plt.figure(figsize=(10, 5))
plt.scatter(correlations_seafood.keys(), correlations_seafood.values(), color='blue', label='Seafood')
plt.title('Correlation with Calories for Seafood')
plt.xlabel('Nutrient')
plt.ylabel('Correlation with Calories')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

```
